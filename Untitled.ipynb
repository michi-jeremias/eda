```python
from eda import DataMangler
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("white")
```

```python
class DataMangler:
    def __init__(self, data):
        self.data = data if isinstance(data, pd.DataFrame) else pd.DataFrame(data)

    def drop(self, col):
        if not isinstance(list, col):
            col = [col]

        try:
            for c in col:
                self.data.drop(c, inplace=True)
        except IndexError:
            print('Invalid index')

    def eda(self):
        print(f"Data:\n\
                Shape: \t{self.data.shape}\n\
                ")
        for col in self.data.columns:
            print(f"'{col}':\n\
                  Num Unique: \t\t{self.data[col].nunique()}\n\
                  First 10 Unique: \t{self.data[col].unique()[:10]}\n\
                  Type: \t\t{self.data[col].dtype}\n\
                  Missing total \t{self.data[col].isnull().sum()}\n\
                  Missing relative: \t{round(100*self.data[col].isnull().sum()/self.data.shape[0], 2)}%\n\
                  ")
    
    def plot(self):
        for col in self.data.columns:
            # if self.data[col].dtype == 'int64' or 'float64':
            # if self.data[col].dtype == 'float64':
            sns.histplot(self.data[col], color="dodgerblue", label="Compact")          
                
```

```python
dm = DataMangler(dftrain)
```
